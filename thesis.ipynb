## Generating and typesetting tables

Tables in reproducible reports should be generated directly from data rather than edited manually. This makes the result easier to update, check, and reuse. In Quarto, Python code can create a table while chunk options decide whether the code is visible in the rendered report or only the output is shown.

For a report or thesis, the priority is readability and consistency: meaningful headers, units, controlled precision, clear alignment, and compact styling. For slides, tables should usually be even simpler, with fewer rows and columns.

### Raw data and formatted output

The first table shows the raw data. It is reproducible, but not yet suitable for a report: column names are technical, the precision is inconsistent, and scientific notation is not reader-friendly.

```python
#| echo: true

import pandas as pd

raw = pd.DataFrame({
    "sample": ["A", "B", "C"],
    "mass_kg": [1.23456, 1234.567, 0.004321],
    "time_s": [0.0123, 15.678, 1234.5],
    "uncertainty": [0.0001234, 12.345, 0.000009876],
    "comment": [
        "reference sample",
        "large value; check calibration",
        "small signal near detection limit"
    ],
})

raw
```

The same data can be formatted before display. Headers include units, numbers use consistent precision, and values such as `1.23e+03` are written as mathematical scientific notation, for example `1.23* 10^3`. In HTML output, right alignment is a practical approximation of decimal alignment.

```python
#| echo: true

def scientific_latex(x, digits=2):
    """Return a number in LaTeX scientific notation."""
    mantissa, exponent = f"{x:.{digits}e}".split("e")
    return rf"{float(mantissa):.{digits}f}* 10^{{{int(exponent)}}}"

formatted = pd.DataFrame({
    "Sample": raw["sample"],
    "Mass (kg)": raw["mass_kg"].map(lambda x: f"{x:.3g}"),
    "Time (s)": raw["time_s"].map(lambda x: f"{x:.2f}"),
    "Uncertainty": raw["uncertainty"].map(scientific_latex),
    "Comment": raw["comment"],
})

formatted
```

This example covers the most common report-table operations: renaming columns, adding units to headers, limiting digits, replacing raw scientific notation, and separating data generation from display formatting. The raw data remain numeric, while the presentation table is made for reading.

### Styling, wrapping, and grouped headers

More advanced formatting should be used sparingly. The example below keeps all styling in one place: numeric columns are aligned to the right, text is aligned to the left, long comments are wrapped, and related columns are visually grouped. The HTML styling is useful for web output; for PDF output, simpler styling is usually more robust.

```python
#| echo: true

display_table = formatted.copy()
display_table["Comment"] = display_table["Comment"].str.replace("; ", ";<br>", regex=False)

display_table.columns = pd.MultiIndex.from_tuples([
    ("Sample", ""),
    ("Measurement", "Mass (kg)"),
    ("Measurement", "Time (s)"),
    ("Measurement", "Uncertainty"),
    ("Notes", "Comment"),
])

display_table.style \
    .hide(axis="index") \
    .set_properties(
        subset=[("Measurement", "Mass (kg)"),
                ("Measurement", "Time (s)"),
                ("Measurement", "Uncertainty")],
        **{"text-align": "right"}
    ) \
    .set_properties(
        subset=[("Sample", ""), ("Notes", "Comment")],
        **{"text-align": "left", "white-space": "normal"}
    ) \
    .set_table_styles([
        {"selector": "th", "props": [("text-align", "center")]},
        {"selector": "td", "props": [("padding", "4px 8px")]},
    ])
```

Cells can also be split or merged conceptually by changing the table structure before display. In pandas, grouped headers are usually created with a `MultiIndex`; repeated labels can be removed, combined, or separated into additional columns depending on what is clearest for the reader. In Quarto reports, this approach is preferable to manually editing table cells after rendering.

In summary, good table generation in Python means keeping the data reproducible and applying formatting only at the final presentation step. A compact workflow is: create a DataFrame, clean the column labels, format numbers, add units, then apply only the styling needed for the target output.

## Why vector graphics are the only appropriate choice for non-photographic figures


```python
# | include: false
import requests

URL = 'https://upload.wikimedia.org/wikipedia/commons/6/6b/Bitmap_VS_SVG.svg'
with open('Bitmap_VS_SVG.svg', 'wb') as f:
    f.write(
        requests.get(
            URL, headers={'User-Agent': 'Mozilla/5.0'}, stream=True
        ).content
    );
```

Vector graphics are the preferred format for plots, diagrams, maps, charts, and most visualisations 
  in technical and professional documents.
Unlike raster graphics (e.g., PNG or JPG), which are composed of pixels, vector graphics (e.g., SVG or PDF) describe 
  shapes mathematically and represent text as structured elements.
As a result, they are resolution-independent and can be scaled without loss of quality.

Screenshots and raster exports (e.g., PNG or JPG) should be avoided for figures whenever 
  a vector format is available.
They should only be used for snapshots of software interfaces or content that cannot reasonably 
  be represented in vector form. 
Raster formats are primarily suitable for photographic images or other inherently pixel-based data.
Even for raster images, all text (labels, titles, annotations) should be added as vector graphics
  by embedding the images in SVG or PDF files which may contain a mixture of both vector and raster content. 
Raster exports of plots, maps and diagrams introduce loss of scalability, non-selectable text, 
  and reduced clarity when printed or zoomed.
  
Vector graphics are the expected format in academic journals and a best practice for professional materials 
  across industries, including logo design and visual identity systems.
SVG is a widely supported vector format viewable in all modern web browsers and editable using open-source 
  tools such as Inkscape, making it highly portable and interoperable.

## Basic plotting with \techword{Matplotlib}


```python
# | label: fig-sin
# | fig-cap: "Sinusoid & co."

import numpy as np
from matplotlib import pyplot

pyplot.plot(
    x := np.linspace(0, 2 * np.pi, 100), np.sin(x), label=r"$\sin(x)$"
)
pyplot.plot(x, np.cos(x / 2), label=r"$\cos(x/2)$")
pyplot.plot(
    x, np.sin(2 * x) * np.cos(x), label=r"$\sin(2x) \cdot \cos(x)$"
)
pyplot.legend();
```
