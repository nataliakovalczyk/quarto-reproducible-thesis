## Generating and typesetting tables

Tables in Quarto can be written directly in **Markdown**, generated from **Python**, or typeset using **LaTeX** and **HTML/CSS** for finer control. A useful distinction is between **generating** a table (creating it from data) and **typesetting** a table (controlling how it is displayed). In practice, data should usually be generated automatically from code and then formatted appropriately for the target output.

For simple tables, Markdown syntax is often sufficient:

```markdown
| Quantity | Value | Unit |
|---|---:|---|
| Mass | 1.23 | kg |
| Time | 0.015 | s |
| Energy | 1.23e+03 | J |
```

Markdown tables are easy to write and portable across formats, but offer limited control over alignment, text wrapping, and numerical formatting.

For computed results, tables should generally be generated from Python (e.g. with `pandas`), ensuring that the displayed values remain consistent with the underlying calculations:

```python
import pandas as pd

df = pd.DataFrame({
    "mass / kg": [1.234, 12.345, 1234.5],
    "time / s": [0.01523, 1.2, 3600],
    "energy / J": [12.3456, 1234.56, 1.23e6],
})

df
```

Raw output is useful for checking calculations, but formatted tables are usually better for communication. For example, compare

```text
mass     time      energy
1.234    0.01523   12.3456
12.345   1.2       1234.56
1234.5   3600      1.23e+06
```

with a formatted version:

```latex
\begin{tabular}{S S S}
\toprule
{Mass / kg} & {Time / s} & {Energy / J} \\
\midrule
1.23 & 0.0152 & 12.3 \\
12.3 & 1.20 & 1.23 \times 10^{3} \\
1.23 \times 10^{3} & 3.60 \times 10^{3} & 1.23 \times 10^{6} \\
\bottomrule
\end{tabular}
```

Several formatting choices improve readability:

* Align numerical columns consistently, ideally at the decimal separator.
* Round values to an appropriate number of **significant digits** rather than displaying all available digits.
* Format scientific notation, e.g. convert `1.23e+03` into $1.23\times10^3$.
* Display time values in a suitable form, such as `15.2 ms`, `1.20 s`, or `01:00:00`.
* Separate quantities from units by placing units in the column header rather than repeating them in each cell.

For example, decimal alignment makes it easier to compare values:

```text
Poor:
1.2
12.34
123.456

Better:
  1.2
 12.34
123.456
```

For PDF output, the LaTeX package `siunitx` is particularly useful because it can automatically align numbers, round values, control decimal markers, and format scientific notation:

```latex
\num{1.23e3}
```

which renders as $1.23\times10^3$.

Headers deserve special attention. They should identify both the quantity and its unit, for example `Mass / kg`, `Length / m`, or `Velocity / m s$^{-1}$`. This keeps the table compact and avoids repeating units in every row.

Complex headers can be created by merging or splitting cells. In LaTeX this is typically done with `\multicolumn` and the `multirow` package:

```latex
\multicolumn{2}{c}{Measurement}
```

Long text inside a cell may require wrapping across multiple lines. Common solutions include fixed-width columns (`p{...}`), `tabularx`, or the `makecell` package. For example:

```latex
\begin{tabular}{p{3cm} S}
```

creates a fixed-width text column that wraps automatically.

Formatting requirements may also depend on the output medium. Tables intended for reports can contain more detail and precise numerical alignment, whereas tables intended for slides should usually contain fewer columns, larger text, and reduced numerical precision.

Finally, Quarto supports additional styling through HTML and CSS. While LaTeX generally provides the highest-quality PDF tables, HTML styling can be useful for web pages and slides, allowing custom colours, fonts, borders, row highlighting, responsive layouts, and interactive elements.

In summary, use Markdown for small manually written tables, Python for reproducible data generation, LaTeX (especially `siunitx`) for precise numerical typesetting in reports, and HTML/CSS when richer web-based styling is required.

## Why vector graphics are the only appropriate choice for non-photographic figures


````python
# | include: false
import requests

URL = 'https://upload.wikimedia.org/wikipedia/commons/6/6b/Bitmap_VS_SVG.svg'
with open('Bitmap_VS_SVG.svg', 'wb') as f:
    f.write(
        requests.get(
            URL, headers={'User-Agent': 'Mozilla/5.0'}, stream=True
        ).content
    );
````

Vector graphics are the preferred format for plots, diagrams, maps, charts, and most visualisations 
  in technical and professional documents.
Unlike raster graphics (e.g., PNG or JPG), which are composed of pixels, vector graphics (e.g., SVG or PDF) describe 
  shapes mathematically and represent text as structured elements.
As a result, they are resolution-independent and can be scaled without loss of quality.

Screenshots and raster exports (e.g., PNG or JPG) should be avoided for figures whenever 
  a vector format is available.
They should only be used for snapshots of software interfaces or content that cannot reasonably 
  be represented in vector form. 
Raster formats are primarily suitable for photographic images or other inherently pixel-based data.
Even for raster images, all text (labels, titles, annotations) should be added as vector graphics
  by embedding the images in SVG or PDF files which may contain a mixture of both vector and raster content. 
Raster exports of plots, maps and diagrams introduce loss of scalability, non-selectable text, 
  and reduced clarity when printed or zoomed.
  
Vector graphics are the expected format in academic journals and a best practice for professional materials 
  across industries, including logo design and visual identity systems.
SVG is a widely supported vector format viewable in all modern web browsers and editable using open-source 
  tools such as Inkscape, making it highly portable and interoperable.

## Basic plotting with \techword{Matplotlib}


````python
# | label: fig-sin
# | fig-cap: "Sinusoid & co."

import numpy as np
from matplotlib import pyplot

pyplot.plot(
    x := np.linspace(0, 2 * np.pi, 100), np.sin(x), label=r"$\sin(x)$"
)
pyplot.plot(x, np.cos(x / 2), label=r"$\cos(x/2)$")
pyplot.plot(
    x, np.sin(2 * x) * np.cos(x), label=r"$\sin(2x) \cdot \cos(x)$"
)
pyplot.legend();
````
